# Limpieza de datos

```python
#Generar una copia del DF para trabajar de forma segura
df_clean = df.copy()
```

```python
#Consultar la cantidad de registros duplicados
df_clean.duplicated().sum()
```

```python
#Eliminar las columnas mencionadas en hallazgos iniciales
df_clean.drop(['Código ISO del país', 'Nombre del país', 'Nombre del grupo étnico'], axis=1, inplace=True)
```

```python
#Comprobación de la eliminación de las columnas
display(df.shape)
df_clean.shape
```

```python
#Verificación tipos de formatos de fecha
df_clean[[
    'fecha reporte web',
    'Fecha de notificación',
    'Fecha de inicio de síntomas',
    'Fecha de muerte',
    'Fecha de diagnóstico',
    'Fecha de recuperación'
    ]]
```

```python
#Conversión a tipo fecha(datetime)
columnas_fecha = [
    'fecha reporte web',
    'Fecha de notificación',
    'Fecha de inicio de síntomas',
    'Fecha de muerte',
    'Fecha de diagnóstico',
    'Fecha de recuperación'
    ]

df_clean[columnas_fecha] = df_clean[columnas_fecha].apply(pd.to_datetime, errors='coerce')
```

```python
#Observación del nuevo formato de fecha
df_clean[[
    'fecha reporte web',
    'Fecha de notificación',
    'Fecha de inicio de síntomas',
    'Fecha de muerte',
    'Fecha de diagnóstico',
    'Fecha de recuperación'
    ]]
```

```python
#Comprobación del nuevo tipo de dato
df_clean.dtypes
```

```python
#Detección de posibles valores duplicados
df_clean.columns
revision_valores_unicos_columnas = ["Sexo", "Ubicación del caso", "Estado", "Recuperado", "Tipo de contagio", 'Tipo de recuperación']

for c in revision_valores_unicos_columnas:
    print(f"{c}: {df_clean[c].unique()}")
```

```python
#Normalización de valores
# Sexo: ['F' 'M' 'f' 'm']
# Ubicación del caso: ['Casa' 'Fallecido' nan 'casa' 'CASA' 'Hospital' 'Hospital UCI']
# Estado: ['Leve' 'Fallecido' nan 'leve' 'LEVE' 'Moderado' 'Grave']
# Recuperado: ['Recuperado' 'Fallecido' nan 'fallecido' 'Activo']
# Tipo de contagio: ['Comunitaria' 'Relacionado' 'Importado']
# Tipo de recuperación: ['Tiempo' nan 'PCR' 'TIEMPO']

reemplazos = {
    'Sexo': {
        'f': 'F',
        'm': 'M'
        },
    'Ubicación del caso': {
        'casa': 'Casa',
        'fallecido': 'Fallecido',
        'hospital': 'Hospital',
        'hospital uci': 'Hospital UCI'
    },
    'Estado': {
        'leve': 'Leve',
        'fallecido': 'Fallecido',
        'moderado': 'Moderado',
        'grave': 'Grave'
    },
    'Recuperado': {
        'recuperado': 'Recuperado',
        'fallecido': 'Fallecido',
        'activo': 'Activo'
    },
    'Tipo de recuperación': {
        'tiempo': 'Tiempo',
        'pcr': 'PCR'
    }
}

for key, value in reemplazos.items():
#items() regresa una tupla de clave:valor a diferencia de values que regresa solo los valores y keys que solo me da la clave
    df_clean[key] = df_clean[key].str.strip().str.lower().replace(value, regex=False)
```

```python
# validación corrección de los valores unicos
revision_valores_unicos_columnas = ["Sexo", "Ubicación del caso", "Estado", "Recuperado", "Tipo de contagio", 'Tipo de recuperación']

for c in revision_valores_unicos_columnas:
    print(f"{c}: {df_clean[c].unique()}")
```

```python
#Validación de nullos
df_clean.isnull().sum()
```
